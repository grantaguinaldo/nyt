```python
import pandas as pd
import requests as r
import json
```

### NYT API

```python
app_key = ''
key = ''
secret = ''
```

```python
query = 'natural%20gas'
base = 'https://api.nytimes.com/svc/search/v2/articlesearch.json?q=' 
data_json = []
for each in range(0, 9, 1):
    end_point = base + query + '&page=' + str(each) + '&api-key=' + key
    data = r.get(end_point).json()
    for index, j in enumerate(data['response']['docs']):
        data_json.append({'url': j['web_url'],'lead': j['lead_paragraph']})
        print('Index: {}: {}'.format(index, j['web_url']))

with open('nytdata.json', 'w') as f:
    json.dump(data_json, f)
```

```python
df = pd.read_json('nytdata.json', orient='columns')
```

### News API

```python
key = ''
base = 'https://newsapi.org/v2/everything?q='
query = 'natural%20gas'
conn_string =  base + query + '&apiKey=' + key + '&pageSize=100' + '&sortBy=popularity'
data = r.get(conn_string).json()
```

```python
data_json_news_api = []
for i, each in enumerate(data['articles']):
    data_json_news_api.append({'url': each['url'],'lead': each['description']})
    print('Index: {}: {}'.format(i, each['url']))
    print('---')
data_json_news_api
```

```python
with open('news_api_data.json', 'w') as f:
    json.dump(data_json_news_api, f)
```

```python
df_news_api = pd.read_json('news_api_data.json', orient='columns')
```
